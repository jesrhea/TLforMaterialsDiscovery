```python
import pandas as pd
import matplotlib.pyplot as plt
high_gas_all = pd.read_csv('mercardo_ch4_high-p_65.csv', names=['name', 'gas-ads'])
plt.hist(
    high_gas_all['gas-ads'],
    log=True, 
    ec='k',
    histtype='bar',
    bins=[x*10 for x in range(15)]
)
#plt.xticks([x*15 for x in range(10)])
plt.xlabel('CH4 Gas Adsorption - 65 bar')
plt.ylabel('Number of COFs')
plt.show() 

high_gas_all_sort = high_gas_all.sort_values('gas-ads')

bins = [x*10 for x in range(15)]
binned_dfs = []
for i,bin_ in enumerate(bins[:-1]):
    df = high_gas_all_sort[
        (bin_<=high_gas_all_sort['gas-ads']) & (high_gas_all_sort['gas-ads']<bins[i+1])]
    binned_dfs.append(df)
    df = None
    
strat_subset = []
sum_of_whole_dataset = len(high_gas_all.index)
print('sum of whole dataset: ', sum_of_whole_dataset)
for df in binned_dfs:
    print('Bin length : ',int(round(len(df.index))) )
    ratio = (len(df.index) / sum_of_whole_dataset)
    print('Ratio : ', ratio)
    subset_num = int(round(ratio * 10000))
    #print('Subset length : ', subset_num,'\n')
    print()
    subset = df.sample(subset_num)
    print('Subset length : ', len(subset.index))
    print('Subset mean : ', subset.loc[:,'gas-ads'].mean(),'\n\n')
    strat_subset.append(subset)
    high_rep_subset = pd.concat(strat_subset)
    
```

```python


import pandas as pd
import matplotlib.pyplot as plt
low_gas_all = pd.read_csv('mercardo_ch4_low-p_5.8.csv', names=['name', 'gas-ads'])
plt.hist(
    low_gas_all['gas-ads'],
    log=True, 
    ec='k',
    histtype='bar',
    bins=[x for x in range(15)]
)
#plt.xticks([x*15 for x in range(10)])
plt.xlabel('CH4 Gas Adsorption - 5.8 bar')
plt.ylabel('Number of COFs')
plt.show() 

low_gas_all_sort = low_gas_all.sort_values('gas-ads')

bins = [x for x in range(15)]
binned_dfs = []
for i,bin_ in enumerate(bins[:-1]):
    df = low_gas_all_sort[(bin_<=low_gas_all_sort['gas-ads']) & (low_gas_all_sort['gas-ads']<bins[i+1])]
    binned_dfs.append(df)
    df = None
    
strat_subset = []
sum_of_whole_dataset = len(low_gas_all.index)
for df in binned_dfs:
    print('Bin length : ',int(round(len(df.index))) )
    ratio = (len(df.index) / sum_of_whole_dataset)
    print('Ratio : ', ratio)
    subset_num = int(round(ratio * 10000))
    #print('Subset length : ', subset_num,'\n')
    print()
    subset = df.sample(subset_num)
    print('Subset length : ', len(subset.index))
    print('Subset mean : ', subset.loc[:,'gas-ads'].mean(),'\n\n')
    strat_subset.append(subset)
    low_rep_subset = pd.concat(strat_subset)
    
```

```python
low_rep_subset.to_csv('mercado_ch4_lowP58_repsubset_10000-2.csv',index=False, header=False, index_label=False)
```
